```python
import sqlite3
import pandas as pd
```

```python
con = sqlite3.connect('nba_salary.sqlite')
```

```python
df = pd.read_sql_query("SELECT * FROM NBA_season1718_salary WHERE season17_18>10000000", con)
```

```python
sql = "SELECT NBA_season1718_salary.*, Seasons_Stats.* FROM NBA_season1718_salary LEFT JOIN Seasons_Stats ON Seasons_Stats.x1 = NBA_season1718_salary.x1 WHERE season17_18 > 10000000"
df2 = pd.read_sql_query(sql, con)
print(df2.shape)
df2.head()
```

```python
sql = "SELECT NBA_season1718_salary.*, Seasons_Stats.* FROM NBA_season1718_salary LEFT JOIN Seasons_Stats ON Seasons_Stats.x1 = NBA_season1718_salary.x1 WHERE season17_18 > 10000000"
df = pd.read_sql_query(sql, con)
df = df.loc[:,~df.columns.duplicated()].copy()
print(df.shape)
df.head()
```

```python
df2 = df[["PTS","AST", "PF"]].mean()
df2
```

```python
# Cálculo de promedios con SQL
sql = "SELECT AVG(Seasons_Stats.PTS) AS 'prom_total_pts', AVG(Seasons_Stats.AST) AS 'prom_total_ast', AVG(Seasons_Stats.PF) AS 'prom_total_pf' FROM NBA_season1718_salary LEFT JOIN Seasons_Stats ON Seasons_Stats.x1 = NBA_season1718_salary.x1 WHERE season17_18 > 10000000"
df = pd.read_sql_query(sql, con)
print(df.shape)
df.head()
```
