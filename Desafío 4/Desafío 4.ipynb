#### 2. Realizar un histograma con los salarios. Qué rango(s) de salarios son los más populares?


```python
#(pag37)se construyen con ax.hist
#bins representa la cantidad de intervalos a construir
```

```python
#pag 66, para ver histogramas con seaborn. 
#Con el parámetro kind='hist', Seaborn graficará un histograma.
```

```python
#sns.set()
df_ejercicio = sns.load_dataset('exercise')
df_ejercicio = df_ejercicio.drop('Unnamed: 0', axis='columns')
df_ejercicio.head()
```

```python
df_30_min = df_ejercicio[df_ejercicio['time'] == '30 min']
df_30_min.head()
```

```python
ax = sns.displot(data=df_30_min, x='pulse', kind='hist', hue='diet', fill=True)
ax.set(xlabel='Frecuencia Cardíaca', ylabel='Densidad', title='Distribución  de las pulsaciones')
plt.figure()
```

```python
ax = sns.displot(data=archivo_csv, x='Salary', kind='hist',bins=15, fill=True)
ax.set(xlabel='Salarios', ylabel='Densidad', title='Distribución  de Salarios')
plt.figure()
```

#### Los salarios más frecuentes están entre 50.000 y 100.000

#### 3 Realizar un gráfico de violín con los salarios, discriminados por género. 


```python
#https://seaborn.pydata.org/generated/seaborn.violinplot.html
#https://www.youtube.com/watch?v=PNNLefP974M
```

```python
ax = sns.catplot(data=archivo_csv,kind='violin', y='Salary',x="GenderID",  split=True)
ax.set(ylabel='Salario', xlabel='Género', title='Distribución  de Salarios según género')
plt.figure()
```
