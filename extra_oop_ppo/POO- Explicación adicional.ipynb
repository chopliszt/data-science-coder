```python
Como todos los botones tienen una informaciÃ³n muy similar podemos llamar a cada uno un "Objeto". Cada persona es un "Objeto" 
que tiene la misma informaciÃ³n. Los objetos necesitan dos cosas: 1)lo que tienen y 2)lo que hacen. A lo primero lo llamamos 
"atributos" y a lo que hacen, lo llamamos "metodos"
```

Entonces, en vez de crear objeto por objeto, vamos a crer una fÃ¡brica de objetos. En Python, una fÃ¡brica de objetos es llamada CLASE

```python
Vamos ahora a crear la fÃ¡brica de botones!
```

```python
#Recordemos la informaciÃ³n de cada botones. Por ejemplo para Lara tenemos:
botones_5_edad=26
botones_5_nombre= "Lara"
botones_5_apellido= "Seufert"
botones_5_idiomas= ["InglÃ©s", "EspaÃ±ol", "AlemÃ¡n"]
botones_5_aÃ±os_de_experiencia= 4
botones_5_tiene_permido_de_trabajar= True
```

```python
#en Python
class Botones :
    def __init__(self,edad,nombre,apellido,idiomas,experiencia,tiene_permido_de_trabajar):
        self.edad=edad
        self.nombre=nombre
        self.apellido=apellido
        self.idiomas=idiomas
        self.experiencia=experiencia
        self.tiene_permido_de_trabajar=tiene_permido_de_trabajar
        
    def saludar_en_esp(self):
        print(f"hola mi nombre es {self.nombre} y te doy la bienvenida al hotel")
    def pedir_propina(self):
        print(f"hola querido huesped, dado mi excelente servicio, puedes darme 200USD de propina")
        
```

```python
#Usando neustra fabrica de Objetos Botones, vamos a crear neustro primer botones. Es asÃ­ de fÃ¡cil y rÃ¡pido
botones_1=Botones(19,"Juan","Infante",["Italiano"],3,True)
```

```python
#Ahora, colocando un punto despuÃ©s del objeto, podemos acceder a toda su informaciÃ³n
print(f"La informaciÃ³n de nuestro primer botones es Edad:{botones_1.edad},Nombre:{botones_1.nombre},Apellido:{botones_1.apellido}")
```

```python
#Y asÃ­ de fÃ¡cil es crear nuestros 5 botones como antes
botones_1=Botones(19,"Felipe","Montes",["InglÃ©s", "EspaÃ±ol"],3,True)
botones_2=Botones(19,"Diego","Fuentes",["Indonesio", "EspaÃ±ol"],3,True)
botones_3=Botones(19,"Camilo","Mindernickel",["InglÃ©s", "EspaÃ±ol", "FrancÃ©s", "AlemÃ¡n", "PortuguÃ©s"],1,True)
botones_4=Botones(19,"Andrea","Campos",["InglÃ©s", "EspaÃ±ol"],0,True)
botones_5=Botones(19,"Gaby","Baudelaire",["InglÃ©s", "EspaÃ±ol", "FrancÃ©s", "AlemÃ¡n", "Indonesio", "JaponÃ©s","Swahili"],2,True)
botones_6=Botones(19,"Lara","Seufert",["InglÃ©s", "EspaÃ±ol", "AlemÃ¡n"],0,True)
```

```python
#tambien podemos pedirle a los botones que saluden, todo de una forma sencilla y rapida
botones_1.saludar_en_esp()
botones_2.saludar_en_esp()
botones_3.saludar_en_esp()
botones_4.saludar_en_esp()
botones_5.saludar_en_esp()
botones_6.saludar_en_esp()
```
