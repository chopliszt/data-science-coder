#### 2 Separar nombres de apellidos en dos columnas distintas usando la coma como separador

```python
temporal=archivo_fuente_juegos["Athlete"].str.split(", ",n = 1, expand = True)
archivo_fuente_juegos["First Name"]=temporal[0]
archivo_fuente_juegos["Last Name"]=temporal[1]
archivo_fuente_juegos.drop(columns =["Athlete"], inplace = True)
archivo_fuente_juegos
```
