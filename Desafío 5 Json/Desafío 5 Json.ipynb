```python
#Importando las librerías externas necesarias para leer el Json
import json
import requests
```

```python
#Importando Pandas
import pandas as pd
```

```python
url="https://jsonplaceholder.typicode.com/users"
```

```python
respuesta=requests.get(url)
respuesta
#respuesta.text
```

```python
JSONData=json.loads(respuesta.text)
```

```python
#Revisamos que si esté trayendo los datos de Json
JSONData
```

```python
#creando el Data Frame desde un Diccionario, porque el JSON es muy parecido a un diccionario de Python
df=pd.DataFrame.from_dict(JSONData)
df
```
